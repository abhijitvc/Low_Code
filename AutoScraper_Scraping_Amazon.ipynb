```python
from autoscraper import AutoScraper
```

```python
url = "https://www.amazon.in/s?k=iphone&ref=nb_sb_noss"
```

```python
wanted_list = ["Apple iPhone 11 (64GB) - White",
              "41,999",
              "72,326"]
```

```python
scraper = AutoScraper()
result = scraper.build(url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_rrik':'Title','rule_mnk9':'Price','rule_o28c':'Ratings'})
scraper.keep_rules(['rule_rrik','rule_mnk9','rule_o28c'])
scraper.save('iphone_search')
```

```python
result = scraper.get_result_similar("https://www.amazon.in/s?k=samsung+s22+ultra&crid=LEGG3W3CP9AQ&sprefix=samsung+s%2Caps%2C206&ref=nb_sb_ss_ts-doa-p_1_9", group_by_alias=True)
```
